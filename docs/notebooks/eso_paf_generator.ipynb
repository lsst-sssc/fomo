```python
target = 'c:40P'
refcenter = '309'  # ESO Paranel
epoch_start = '2026-01-28T00:00:00.0'
epoch_end = '2026-02-04T00:00:00.0'
stepsize = '1'
pafstepsizeunit = 'h'
epoch = epoch_start
epoch += ', ' + epoch_end
epoch += ', ' + stepsize + ' - ' + pafstepsizeunit
paf_params = '2.2, 0'
print(epoch)
```

```python
base_url = 'https://www.eso.org/sci/php/phase2/paf-miriade.php'
params = {
    'paf_body': target,
    'paf_epoch': epoch,
    'paf_refcenter': refcenter,
    'paf_params': paf_params,
}
# Build request to post
resp = requests.post(base_url, params)
if resp.ok:
    print(resp.text)
    # Write to file (needs to strip the leading and trailing <textarea> to be valid PAF)
    with open(f'{target[2:]}_PAF.txt', 'w', encoding='utf-8') as file:
        file.write(resp.text)
else:
    print('Error:', resp.status_code)
```
